```python
# Importation des bibliothèques
import pandas as pd
import matplotlib.pyplot as plt

# Étape 1 : Chargement des données
# Remplace 'passenger_data.csv' par le chemin de ton fichier
file_path = 'passenger_data.csv'
data = pd.read_csv('flights.csv')

# Étape 2 : Création d'une colonne 'date' pour faciliter l'analyse
data['date'] = pd.to_datetime(data['year'].astype(str) + ' ' + data['month'])

# Étape 3 : Regrouper les données par 'date' et sommer les passagers
# Cela permet d'obtenir le trafic global des passagers pour toutes les compagnies
global_trend = data.groupby('date')['passengers'].sum().reset_index()

# Étape 4 : Tracer la courbe de tendance globale
plt.figure(figsize=(12, 6))
plt.plot(global_trend['date'], global_trend['passengers'], color='blue', marker='o', linestyle='-', label='Global Passenger Traffic')
plt.title('Tendance Globale Mensuelle du Trafic des Passagers')
plt.xlabel('Date')
plt.ylabel('Nombre de Passagers')
plt.legend()
plt.grid(True)
plt.show()
```
